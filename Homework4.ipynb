```python
#5
marks = {'Andy':88, 'Amy':66, 'James': 90, 'Jules': 55, 'Arthur': 77}
x= np.array(['Andy','Amy','James','Jules','Arthur'])
y= np.array([88,66,90,55,77])
plt.bar(x,y)
plt.title("Student's Grades")
plt.show()
```

```python
mylabels= ['Andy','Amy','James','Jules','Arthur']
y= np.array([88,66,90,55,77])
plt.pie(y, labels= mylabels)
plt.title("Student's Grades")
plt.legend(bbox_to_anchor= (1.05,1.0), loc='upper left')
plt.show()
```

```python
#6
import matplotlib.pyplot as plt
import numpy as np
x = np.array([0, 1, 2, 3])
y = np.array([0, 4, 10, 12])
x2= np.array([0, 1, 2, 3])
y2 = np.array([10, 3, 5, 1])
plt.subplot(2, 3, 1)
plt.plot(x,y)
plt.plot(x2,y2)

z= np.array(["A", "C", "T", "S"])
a= np.array([1,2,3,10])
plt.subplot(2, 3, 2)
plt.bar(z,a)
plt.show()

b= np.array([10,20,30,40,50])
plt.subplot(2, 3, 3)
plt.pie(b)
plt.show()

c= np.array([0, 1, 2, 3])
y3 = np.array([0, 1, 2, 5])
plt.subplot(2, 3, 4)
plt.plot(c,y3)
plt.grid()
plt.show()

d= np.array([0, 4, 6, 10])
e = np.array([0, 7, 3, 6])
plt.subplot(2, 3, 5)
plt.scatter(d,e, color= "green" )
plt.show()


f= [1,2,3,5,9,5,3,8]
plt.subplot(2, 3, 6)
plt.hist(f)
plt.show()
```
