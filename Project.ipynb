```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import json
import statistics

#Question 1
df = open("numbers.txt", "r")
df= df.read()
list1 = df.split(" ")
list2 = [item.replace('\n', ',') for item in list1]
for i in list2:
    new= i.split(',')
print(new)
```

```python
freq = {}
for items in new:
    freq[items] = new.count(items)
for key, value in freq.items():
    print (key, ":", value, "times")
```

```python
plt.bar(list(freq.keys()), freq.values(),color='b')
plt.show()
```

```python
with open('freq.json', 'w') as out:
    json.dump(freq, out)
```

```python
#Question 2
df= pd.read_csv('amazon.csv')
df.head()
```

```python
df = df.fillna(0)
df["Total Charged"] = df["Total Charged"].str.replace('$','').astype(float)
df.head(10)
```

```python
sum = df["Total Charged"].sum()
print("The total spent from March 2019 to March 2022 is $",sum)
```

```python
print("Average amount spent per order is $",  '%.2f' % df["Total Charged"].mean())
print("Median amount spent per order is $",  '%.2f' % df["Total Charged"].median())
print("Maximum amount spent per order is $",  '%.2f' % df["Total Charged"].max())
print("Minimum amount spent per order is $",  '%.2f' % df["Total Charged"].min())
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
df.head()
```

```python
df.plot.bar(x='Order Date', y='Total Charged', rot=90, figsize=(20,10))
```

```python
mylabels= df['Total Charged']
y= np.array(df['Total Charged'])

plt.pie(y, labels=mylabels)
plt.legend(bbox_to_anchor=(1.05,1.0), loc= 'upper left')
plt.show()
```
